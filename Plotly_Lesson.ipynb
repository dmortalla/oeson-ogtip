```python
import plotly.express as px
import pandas as pd

df = px.data.gapminder()
df.head()

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/gapminder2007.csv')
df
```

```python
# prompt: Using dataframe df: year vs GDP per

# There is no year column in the dataframe.
# Please update the dataframe or specify which column to use for the x-axis.
```

```python
# Filter data
df_us = df[df['country'] == 'United States']

# Create a line chart
fig = px.line(df_us, x='year', y='gdpPercap', title='GDP Per Capita by Year')

# Show the chart
fig.show()
```
